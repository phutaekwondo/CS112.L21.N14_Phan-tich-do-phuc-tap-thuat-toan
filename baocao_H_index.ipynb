##Complexity

- Hàm sort() của python sử dụng thuật toán Timsort có thời gian chạy T1 = nlog(n) * C1


- Sau đó chạy vòng lặp, vòng lặp này có thể lặp n lần nếu gặp "Trường hợp xấu nhất" => đoạn vòng lặp có thời gian chạy T2 = n * C2

Vậy thời gian chạy cả 2 phần tính theo quy tắc cộng T = T1 + T2 = nlog(n)*C1 + n*C2

==>>   Độ phức tạp của thuật toán là O(nlog(n))

##Programming

```python
n = int(input())
arr = [int(x) for x in input().split()]

arr.sort()

k = len(arr)
i = 0

while i < n:
    if arr[i] >= k:
        i = n

    if i != n :
        i += 1
        k -= 1

print(k)
```
