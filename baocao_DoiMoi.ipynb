# Programming

```python
a,k,b,m,n = [ int( x ) for x in input().split() ]


day = 0 
daycur = 0
while n > 0 :
    daycur = n//( a+b )
    if daycur == 0:
        daycur = 1

    n -= daycur*( a+b ) 
    n += a*( ( day + daycur)//k - ( day//k ) ) 
    n += b*( ( day + daycur)//m - ( day//m ) )     
    day += daycur

print(day)
```
