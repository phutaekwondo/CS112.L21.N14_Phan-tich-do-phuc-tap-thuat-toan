# Complexity

Ý tưởng của thuật toán:

+ khởi tạo #day = 0, biến này sẽ là output
+ Chọn số ngày làm việc #daycur ( nếu cây có thể chặt hết trong 1 ngày thì #daycur = 1 ). #daycur được tính bằng công thức: | #daycur = n//(a+b) |
+ Tính số cây còn lại sau #daycur ngày làm việc, số cây còn lại tức là số cây không chặt trong các ngày nghỉ. có thể tính như sau: 

        số cây dư sau khi chặt đúng #daycur ngày không nghỉ = n - daycur*(a+b)
                                                            = n -[n//(a+b)]*(a+b)
                                                            ~= 0 (xấp sỉ)
        số cây không chặt trong các ngày nghỉ = a*[ (day+dayur)//k - (day//k) ] 
                                              + b*[ (day+daycur)//m - (day//m) ]
                                              ~= a * daycur/k + b * daycur/m (xấp sỉ)
    
    vậy số cây còn lại là tổng của kết quả 2 phép tính trên
        số cây còn lại xấp sỉ = a * daycur/k + b * daycur/m
                              = ( a/k + b/m ) * daycur
                              = ( am + bk )/km * n / (a+b)
                              = n / [ (a+b)km / (am + bk) ]
        **note: khi dùng giá trị xấp sỉ thì sẽ có chênh lệch một ít với thời gian chạy thực tế của thuật toán, nhưng không ảnh hưởng khi tính toán độ phức tạp
+ Dựa vào số cây còn lại để tiếp tục tính toán số ngày làm việc
+ Thuật toán kết thúc khi số cây còn lại <= 0

# Programming

```python
a,k,b,m,n = [ int( x ) for x in input().split() ]


day = 0 
daycur = 0
while n > 0 :
    daycur = n//( a+b )
    if daycur == 0:
        daycur = 1

    n -= daycur*( a+b ) 
    n += a*( ( day + daycur)//k - ( day//k ) ) 
    n += b*( ( day + daycur)//m - ( day//m ) )     
    day += daycur

print(day)
```
