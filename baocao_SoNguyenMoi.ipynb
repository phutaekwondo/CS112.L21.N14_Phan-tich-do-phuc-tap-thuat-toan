# **Complexity**

Ý tưởng: list_num dùng để truy cập vào list với các số dư khi chia cho 3. Ví dụ: a%3=1, 2, 0 thì ta sẽ truy cập vào các list tương ứng là ['8','5','2'], ['7','4','1'], ['9','6','3'].

kiểm tra số tại vị trị index của số:

  +) key = num[index], idx = 2 - tổng các số còn lại trong mảng % 3

  +) Ví dụ tổng vị trí còn lại % 3 == 2, thì ta truy cập vào ['7','4','1'] các số này % 3 == 1. Để khi ta thay 1 trong 3 số này vào thì để số mới có thể chia hết cho 3.

  +) Tìm số lớn nhất và lớn hơn key trong list_num[idx], nếu có thì thay vào vị trí index và end game, không thì xét tiếp.

  +) Coi num[index] là phẩn tử cuối của num không, nếu không thì cho index + 1 và xét lại, nếu đúng thì xét tiếp.
  
  +) Chỉ cần tìm số lớn nhất và khác key trong 3 số thay vào num[index] và end game. Notice: Chắc chắn trong ba số sẽ có số khác key nên trên lưu đồ không có nhánh 'No' tại image.png

Việc truy cập vào list_num chỉ chạy vòng lặp 3 lần, không phụ thuộc vào dữ liệu đầu vào. Và các statement cũng thế. Chúng có T0 = 1.

  +) Ta thấy trong trường hợp xấu nhất là duyệt index đến chữ số cuối cùng nên T1 = C1 * n.

  +) Mỗi lần duyệt là ta tính tổng các chữ số còn lại của num mà sum trong python có độ phức tập là O(n) =) T2 = C2*(n-1)

  +) T = T1 * T2. T0 = 1 khi ta lấy làm tròn thì T0 cũng không là gì so với T1, T2 nên ta có thể khỏi tính vào.

  +) T = C1 * n * C2 * (n-1) = C * n^2 - n.

  +) Đoạn code trên có độ phức tập là O(n^2)

# **Programming**

```python
def new_int(num, index):
    list_num = [['7','4','1'],['8','5','2'],['9','6','3']]
    num_str = list(num)
    num_int = [int(i) for i in num_str]
    key = num_int[index]
    list_idx = 2 - sum(num_int[:index] + num_int[index+1:])%3
    for j in list_num[list_idx]:
        if key < int(j):
            num_str[index] = j
            return ''.join(num_str)
    if index == len(num_str) - 1:
        for j in list_num[list_idx]:
            if key != int(j):
                num_str[index] = j
                return ''.join(num_str)
    return new_int(num, index + 1)

num = input()
print(new_int(num,0))
```
