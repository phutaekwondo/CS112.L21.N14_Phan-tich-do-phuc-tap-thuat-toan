# **Complexity**

Ý tưởng: list_num dùng để truy cập vào list với các số dư khi chia cho 3. Ví dụ: a%3=1, 2, 0 thì ta sẽ truy cập vào các list tương ứng là ['8','5','2'], ['7','4','1'], ['9','6','3'].

kiểm tra số tại vị trị index của số:

  +) key = num[index], sur =tổng các số còn lại trong mảng % 3 bằng công thức:
  sur = _sur[sum(num)%3 - key%3] với _sur=[0,1,2]

  +) Ví dụ tổng vị trí còn lại % 3 == 2, thì ta truy cập vào ['7','4','1'] các số này % 3 == 1. Để khi ta thay 1 trong 3 số này vào thì để số mới có thể chia hết cho 3.

  +) Tìm số lớn nhất và lớn hơn key trong list_num[idx], nếu có thì thay vào vị trí index và end game, không thì xét tiếp.

  +) Coi num[index] là phẩn tử cuối của num không, nếu không thì cho index + 1 và xét tiếp, nếu đúng thì xét tiếp.
  
  +) Chỉ cần tìm số lớn nhất và khác key trong 3 số thay vào num[index] và end game. Notice: Chắc chắn trong ba số sẽ có số khác key.

Việc truy cập vào list_num chỉ chạy vòng lặp 3 lần, không phụ thuộc vào dữ liệu đầu vào. Và các statement cũng thế. Chúng có T0 = 1.

  +) Ta thấy trong trường hợp xấu nhất là duyệt index đến chữ số cuối cùng nên T1 = C1 * n.

  +) Ta tính tổng các chữ số của num mà sum trong python có độ phức tập là O(n) =) T2 = C2*n

  +) T = T1 + T2. T0 = 1 khi ta lấy làm tròn thì T0 cũng không là gì so với T1, T2 nên ta có thể khỏi tính vào.

  +) T = C1 * n + C2 * n = C * n.

  +) Đoạn code trên có độ phức tập là O(n)

# **Programming**

```python
_sur = [0, 1, 2]
list_num = [[9, 6, 3], [8, 5, 2], [7, 4, 1]]
def new_int(num):
    num_int = [int(i) for i in num]
    list_idx = sum(num_int)%3
    for index in range(len(num_int)):
        key = num_int[index]
        sur = _sur[list_idx - (key%3)]
        for j in list_num[sur]:
            if key < j:
                num[index] = j
                return num
        if index == len(num) - 1:
            for j in list_num[sur]:
                if key != j:
                    num[index] = j
                    return num

num = list(input())
result = list(map(str, new_int(num)))
print(''.join(result))
```
