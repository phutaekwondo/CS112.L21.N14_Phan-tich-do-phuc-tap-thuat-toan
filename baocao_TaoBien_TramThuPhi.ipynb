##Tảo biển

###Abstraction

- input: n, k (1<= n <= 1000, 1 <= k <= 10^5)
- Ban đầu : bậc 1 có số lượng n (cá thể)
- Mỗi vòng lặp (lặp k vòng)
	+ x = tổng các (bậc * giá trị)
	+ các bậc đều tăng lên 1 đơn vị
	+ bậc 1 có số lượng x
- output: xuất ra màn hình tổng số lượng của các bậc theo dạng chia cho 10^9+7 lấy phần dư	  



VD: 
n = 3, k =2
ta có : . bậc 1, số lượng n

vòng 1 : 
    x = 1 * n = n
    ta có : 

        . bậc 2: số lượng n
        . bậc 1: số lượng n
vòng 2:
    x = (1 * n) + (2 * n) = 3n
    ta có : 
    
        . bậc 3: số lượng n
        . bậc 2: số lượng n
        . bậc 1: số lượng 3n
Xuất ra tổng số lượng của các bậc là 5n.

###Pattern Recognition



Bài tập dạng tính tổng chuỗi số có quy tắc dùng vòng lặp,
đặc điểm: mỗi lần lặp, các con số tăng lên theo quy tắc cụ thể.

-	Khối lệnh trong hình chữ nhật đỏ : có độ phức tạp về thời gian và không giang đều là O(1)
-	Khổi lệnh trong hình chữ nhật đỏ nằm trong vòng lặp(được rào bởi hình chữ nhật xanh) mà ở đó khối lệnh được lặp lại k lần
-	=> Time(n,k) = C * k
-	=> Space(n,k) = C
-	Độ phức tạp thời gian : O(k)
-	Độ phức tạp không gian : O(1)


###Programming

```python
import sys

input = sys.stdin.readline
output = sys.stdout.write

n,k = [int(x) for x in input().split()]

m = 1000000007

now = n
sum = n
for i in range(k):
    now += sum % m
    sum += now % m
    sum %= m
    now %= m

output(str(now))
```

##Trạm Thu Phí

###Abstraction

Cho một mãng số nguyên gồm n phần tử, tìm tổng giá trị của các phần tử liên tiếp trong mảng sao cho tổng lớn nhất, xuất chỉ số đầu p và chỉ số cuối q của khoảng đó.

###Pattern Recognition



bài tập dạng duyệt mảng, đặc điểm nhận dạng: có tổng số phần tử và các giá trị từng phần tử trong mảng.

Tương tự như bài tập Tảo Biển, khối lệnh trong hộp đỏ có độ phức tạp thời gian và không gian ~ O(1)

Khối lệnh được lặp k lần

=> T(n) = C * n

=> S(n) = C

độ phức tạp thời gian : O(n)

độ phức tạp không gian : O(1)


###Programming

```python
import sys

input = sys.stdin.readline
output = sys.stdout.write

n = int(input())

arr = [int(x) for x in input().split()]

p = q = newp = sum = 0
max =  arr[0]



for index in range(len(arr)):
        sum += arr[index]
        if sum > max:
            q=index
            p = newp
            max = sum
        if sum < 0:
            sum = 0
            newp=index + 1

            
output(str(p+1) + ' ' + str(q+1) + ' ' + str(max))
```
