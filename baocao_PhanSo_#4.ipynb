# **Abstraction**
>Cho 2 phân số _a/b_ và _c/d_. Khi _a/b ≠ c/d_ thì tăng _a_ và _b_ thêm 1, sau đó ước giản phân số _a/b_ vừa nhận được rồi quay lại kiểm tra _a/b_ với _c/d_.

>Tìm số lần biến đổi để _a/b = c/d_, xuất ra 0 nếu không có cách biển đổi để _a/b = c/d_.



# **Pattern Regcognition**

>Bài toán dạng duyệt vòng lặp. Dùng vòng lặp thay đổi dữ liệu để duyệt các trường hợp, trường hợp nào thỏa điều kiện theo đề bài thì sẽ kết thúc vòng lặp.

# **Complexity**
* **Ý tưởng**:

  Khởi tạo một biến _count_ = 0 và vòng lặp while với điều kiện lặp là _a/b != c/d_.

  Khi vào vòng lặp sẽ kiểm tra điều kiện _a/b > c/d_ (nếu đúng xuất ra màn hình giá trị 0 và kết thúc chương trình); tăng _count_, _a_ và _b_ thêm 1, sau đó gán giá trị của _a_, _b_ lần lượt là _a//uoc_, _b//uoc_ (với _uoc_ là ước chung lớn nhất của _a_ và _b_), sau đó quay trở lại đầu vòng lặp.

  Khi vòng lặp kết thúc (_a/b = c/d_), giá trị _count_ sẽ là kết quả cần tìm.
* **Độ phức tạp**:

# **Programming**



```python
import math

a = int(input())
b = int(input())
c = int(input())
d = int(input())

count = 0

while a/b != c/d :
    if ( a/b > c/d ):
        print( 0 )
        exit()   
    
    count += 1

    a += 1
    b += 1
    uoc = math.gcd( a,b ) 
    a = a // uoc 
    b = b // uoc 

print( count )
```
